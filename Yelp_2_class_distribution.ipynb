```
from collections import Counter
import pickle
import matplotlib.pyplot as plt
import numpy as np

#yelp dataset statistics

answers = []
for count in range(80):
  with open(f"yelp_neutral_{count}.txt", 'rb') as f:
    texts = pickle.load(f)
    for txt in texts:
      answers.append(int(txt["stars"]))

star_dist = Counter(answers)
```

```
#Yelp-2 dataset rating statistics
fig, ax = plt.subplots()
x_ax = np.arange(len(list(star_dist.keys())))
ax.bar(x_ax, list(star_dist.values()))
ax.set_xticks(x_ax)
ax.set_xticklabels(list(star_dist.keys()))
ax.set_title("star frequencies (millions)")
plt.show()
```
